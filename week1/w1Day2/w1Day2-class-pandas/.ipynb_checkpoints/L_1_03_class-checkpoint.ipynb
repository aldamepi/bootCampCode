In Pandas, every row and column has an **implicit numeric index** and explicit **labeled index**. We can use either of them to get slices of our dataframe. However, we can't mix both of them. 

* If we want to use the **numeric index**:

    .iloc[list_of_row_indexes/slice_of_row_indexes, list_of_column_indexes/slice_of_column_indexes] (exclusive)

* If we want to use the **label index**:

    .loc[list_of_row_label_indexes, list_of_column_label_indexes] (inclusive)

Most of the time, the row **numeric indexes** will be also **labeled indexes** but they can be different.     



```python
# Working with indexes
#filtered[1:4]
#filtered[['gender', 'ic2', 'ic3']][0:10]
#filtered.loc[0:10,['gender', 'ic2', 'ic3']]
#filtered.loc[1:4]
#filtered.loc[100]
##----These guys are equivalent
display(filtered[1:4])
display(filtered.iloc[1:4])
# But this not
display(filtered.loc[1:4])
```

```python
# now, working just on the indexes row,columns
filtered.iloc[0:10,0:4]
#filtered.iloc[[1,2,4,3,8],[0,2,4]]
```
